```python
import pickle
import numpy
import matplotlib.pylab as plt
plt.style.use("https://raw.githubusercontent.com/NeuromatchAcademy/course-content/master/nma.mplstyle")
```

```python
with open('100_decision_time_0810_15_55.pkl', 'rb') as f:
    rt = pickle.load(f)
    
with open('low_rmp_80mV_pop_rate_0812_16_21.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
rt_np =  numpy.array(rt)
rt_np_sort = rt_np.argsort()
median_index = rt_np_sort[50]


```

```python
def overlap_rate(data, start, end, trial_num):
    time = data[0]['time']

    for i in range(trial_num):
        
        plt.plot(time[start:end], data[i]['pop_rate_a'][start:end], 'c', alpha = 0.1, linewidth= 3)
        plt.plot(time[start:end], data[median_index]['pop_rate_a'][start:end], 'k', alpha = 1, linewidth= 1)
        #plt.plot(time, rate_mean, 'b-', label='mean_1', linewidth=0.8)

        
    plt.ylabel('Firing rate (Spikes/s)')
    plt.xlabel('time (ms)')
    plt.show()
```

```python
# find a mean value 

time = data[0]['time']
```

```python
# make a zero array 

num_trial = len(data)
num_time = len(data[0]['pop_rate_a'])
rate_mat = numpy.zeros([num_trial, num_time])
```

```python
for i in range((num_trial)):
    rate_mat[i,:] = data[i]['pop_rate_a']
    
    
```

```python
numpy.sqrt(100)
```

```python
rate_mean = numpy.mean(rate_mat, axis=0)
rate_sem = numpy.std(rate_mat, axis=0)/numpy.sqrt(num_trial)
```

```python
plt.plot(time, rate_mean, 'b-', label='mean_1', linewidth=0.8)
plt.fill_between(time, rate_mean - rate_sem, rate_mean + rate_sem, color='b', alpha=0.2)
plt.show()
```

```python
overlap_rate(data, 0, 4000, 100)
```
